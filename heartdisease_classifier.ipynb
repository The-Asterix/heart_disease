```python
import pandas as pd
df=pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df1=df[df["Cholesterol"]<=(df["Cholesterol"].mean()+3*df["Cholesterol"].std())]
df1.shape
```

```python
df2=df1[df1["RestingBP"]<=(df1["RestingBP"].mean()+3*df1["RestingBP"].std())]
df2.shape
```

```python
df3=df2[df2["FastingBS"]<=(df2["FastingBS"].mean()+3*df2["FastingBS"].std())]
df3.shape
```

```python
df4=df3[df3["MaxHR"]<=(df3["MaxHR"].mean()+3*df3["MaxHR"].std())]
df4.shape
```

```python
df5=df4[df4["Oldpeak"]<=(df4["Oldpeak"].mean()+3*df4["Oldpeak"].std())]
df5.shape
```

```python
df6=df5[df5["HeartDisease"]<=(df5["HeartDisease"].mean()+3*df5["HeartDisease"].std())]
df6.shape
```

```python
df6.head()
```

```python
df6.ExerciseAngina.unique()
```

```python
df6.ST_Slope.unique()
```

```python
df6.RestingECG.unique()
```

```python
df7=df6.copy()
```

```python
df7.ExerciseAngina.replace(
    {'Y':1,
     'N':0}, inplace=True)
df7.ST_Slope.replace(
    {'Up':0, 'Flat':1, 'Down':2},
    inplace=True)
df7.RestingECG.replace(
    {'Normal':0, 'ST':1, 'LVH':2},
    inplace=True)
```

```python
df7.head()
```

```python
df8=pd.get_dummies(df7, drop_first=True)
```

```python
df8.head()
```

```python
x=df8.drop("HeartDisease", axis=1)
y=df8.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x_scaled=s.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,
                            test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier as rcf
m=rcf()
m.fit(x_train,y_train)
```

```python
m.score(x_test,y_test)
```

#### Use PCA

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
x_pca=pca.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_pca,y,
                            test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier as rcf
m1=rcf()
m1.fit(x_train,y_train)
```

```python
m1.score(x_test,y_test)
```
